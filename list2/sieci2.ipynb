```python
heart_disease = fetch_ucirepo(id=45)

X = heart_disease.data.features
y = heart_disease.data.targets
y = y.replace([2, 3, 4], 1)

X["num"] = y

# deleting rows with nan
X = X.dropna(how='any')

y = X['num']
X = X.drop('num', axis=1)
```

```python
# Transforming categorical values into binary, one-hot encoded values
X = pandas.get_dummies(X, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal'], dtype=float)
# X
```

```python
# define W and b
cols = len(X.axes[1])
W = np.ones(cols)
b = 0
```

```python
# define sigmoid, necesary for output calculation
def sigmoid(z):
    return 1 / (1 + math.e**(-z))
```

```python
def p(x):
    return sigmoid(W*x+b)
```

```python
# loss function
def compute_cost(X, y, W, b):
    m = len(y)
    y_pred = sigmoid((X * W).sum() + b)
    cost = np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
    return cost
```

```python
# train and test sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# # define W and b
# cols = len(X.axes[1])
# W = np.ones(cols)
# b = 0

def gradient_descent(X, y, W, b, learning_rate, num_epochs):
    m = len(y)
    for epoch in range(num_epochs):
        for i in range(m):
            x_i = X.iloc[i].values
            y_pred = sigmoid((x_i * W).sum() + b)

            # if i == 94:
            #     print(x_i.T)
            #     print(y_pred)
            #     print(y.iloc[i])
            #     print((y_pred - y.iloc[i]))
            #     print("dW", dW)
            #     print("db", db)
            #     print("w", W)
            #     print("b", b)
                
            dW = np.dot(x_i.T, (y_pred - y.iloc[i]))
            db = y_pred - y.iloc[i]
            W -= learning_rate * dW
            b -= learning_rate * db
        cost = compute_cost(X, y, W, b)
    return W, b

learning_rate = 0.1
epochs = 100
W, b = gradient_descent(x_train, y_train, W, b, learning_rate, epochs)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def calculate_accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred)

def calculate_precision(y_true, y_pred, average='binary'):
    return precision_score(y_true, y_pred, average=average)

def calculate_fscore(y_true, y_pred, average='binary'):
    return f1_score(y_true, y_pred, average=average)
```

```python
predictions = sigmoid(np.dot(x_test, W) + b)

predicted_values = []
for y in predictions:
    if y <= 0.5:
        predicted_values.append(0)
    else:
        predicted_values.append(1)

print(predicted_values)
```

```python
accuracy = calculate_accuracy(y_test, predicted_values)
precision = calculate_precision(y_test, predicted_values, average='binary')
fscore = calculate_fscore(y_test, predicted_values, average='binary')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"F-Score: {fscore}")
```
